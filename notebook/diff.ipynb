# Diff

```python
import pandas as pd
import numpy as np
```

测试真实数据集

```python
df = pd.read_csv('../jena_climate_2009_2016.csv')
```

```python
series = np.array(df.iloc[:, 2].values)
pd.Series(series).plot.line()
pd.Series(series).diff().plot.line()
```

```python
pd.Series(series).diff().plot.hist(bins=128)
```

Conclusions<br/>
<b>特征服从极值分布</b><br/>
预测误差服从正态分布
