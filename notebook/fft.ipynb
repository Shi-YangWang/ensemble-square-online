# Discrete Fourier Transform

```python
import numpy as np
import pandas as pd
```

```python
# 生成测试数据
x=np.linspace(0,8*2*np.pi,8*1440)
y=0.8 * np.sin(x) + 0.10 * np.random.randn(8 * 1440)
series = pd.Series(y)
```

```python
series.plot.line()
```

```python
transformed = np.fft.fft(series)
```

```python
reconstructed = np.fft.ifft(transformed)
```

```python
pd.Series(np.real(transformed)).plot.hist(bins=128)
```

Conclusion<br/>
<b>FFT实部服从正态分布</b>
