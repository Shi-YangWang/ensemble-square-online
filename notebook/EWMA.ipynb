# Code Repo

```python
'''
    stream detector: EWMA
'''
def ewma(series):
    alpha = 0.2
            # hyper parameters
    n = series.size
            # sample length
    y_lag = 0 # $y_{t-1}$
    y_prd = np.zeros(shape=(n))
            # predictions
    for i in range(0, n):
        y_t = series[i]
            # observation
        y_hat = alpha * y_t + (1-alpha) * y_lag
            # equation 1
        y_lag = y_hat
        y_prd[i] = y_hat
    return y_prd
```
